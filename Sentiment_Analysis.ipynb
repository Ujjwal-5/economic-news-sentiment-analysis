```
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
import time
import pandas as pd
from bs4 import BeautifulSoup
from textblob import TextBlob

options = Options()
options.add_argument('--headless=new')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
options.binary_location = '/usr/bin/google-chrome'

service = Service(ChromeDriverManager().install())

driver = webdriver.Chrome(service=service, options=options)

driver.get("https://www.businesstoday.in/latest/economy")

time.sleep(3)

for i in range(130):
    try:
        load_more = driver.find_element(By.ID, "load_more")

        # Scroll to button
        driver.execute_script("arguments[0].scrollIntoView(true);", load_more)
        time.sleep(1)

        # Click
        driver.execute_script("arguments[0].click();", load_more)

        time.sleep(5)

    except:
        print("No more Load More button")
        break

soup = BeautifulSoup(driver.page_source, "html.parser")

data = []
for link in soup.find_all('a'):
    text = link.string

    if text and len(text.strip()) > 35:
        title = text.strip()

        href = link.get("href")
        if href and href.startswith("/"):
            href = "https://www.businesstoday.in" + href

        # Sentiment Analysis
        polarity = TextBlob(title).sentiment.polarity
        sentiment = "Positive" if polarity > 0 else "Negative" if polarity < 0 else "Neutral"

        data.append([title, href, sentiment])

df = pd.DataFrame(data, columns=["Title", "Link", "Sentiment"])
df = df.drop_duplicates(subset=["Title"])

print(df)
df.to_csv("news_economic_data.csv", index=False)

driver.quit()
```
